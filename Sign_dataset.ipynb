```python
# IMPORTING ALL REQUIRED LIBRARIES# IMPOR
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn import datasets
from sklearn import metrics
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import VotingClassifier
from sklearn import tree
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# LOADING THE DATASET : Sign MNIST
df=pd.read_csv("C:/Users/Mehak Sood/Downloads/sign_mnist_train.csv")
df
```

```python
df.shape
```

```python
# X CONTAINS DATA EXCEPT COLUMNN - "LABEL"
# Y CONTAINS COLUMN - "LABEL"
Y=df['label']
```

```python
del df['label']
X=np.matrix(df)
```

```python
X
```

```python
X.shape
```

```python
# VIEWING THE IMAGE
plt.imshow(X[0].reshape(28,28))
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.3)
X_train.shape,Y_train.shape
```

# Implementing Classifier Models And Also Comparing Different Classifiers 
# 1.Logistic Regression

```python
# IMPLEMENTING CLASSIFIER MODELS AND ALSO COMPARING DIFFERENT CLASSIFIERS
model=LogisticRegression()
model.fit(X_train,Y_train)
```

```python
pred=model.predict(X_test)
pred
```

```python
print(confusion_matrix(Y_test,pred))
```

```python
print(classification_report(Y_test,pred))
```

```python
logistic=metrics.accuracy_score(pred,Y_test)
logistic
```

# 2. Decision Tree Classifier 

```python
model1=tree.DecisionTreeClassifier()
model1.fit(X_train,Y_train)
```

```python
pred1=model1.predict(X_test)
pred1
```

```python
print(confusion_matrix(Y_test,pred1))
```

```python
print(classification_report(Y_test,pred1))
```

```python
tree=metrics.accuracy_score(pred1,Y_test)
tree
```

# 3. Support Vector Machine 

```python
model2=SVC(kernel='poly')
model2.fit(X_train,Y_train)
```

```python
pred2=model2.predict(X_test)
pred2
```

```python
print(confusion_matrix(Y_test,pred2))
```

```python
print(classification_report(Y_test,pred2))
```

```python
SVC=metrics.accuracy_score(pred2,Y_test)
SVC
```
